```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Matplotlib

## Scatter plot

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

## Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
```

## Box-plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

## Pie chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)
plt.axis('equal')
```

# Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

## Scatter plot

```python
sns.scatterplot(x="total_bill",
                    y="tip",
                    data=df)
```

## Histogram

```python
sns.distplot(df["total_bill"])
```

```python
## Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

## Boxplot

```python
sns.boxplot(x="sex",
                y="tip",
                data=df)
```

## Violin plot

```python
sns.violinplot(x="tip", y="sex", data=df,palette='rainbow')
```

## More visualization with Seaborn:

### Heatmap

```python
sns.heatmap(df.corr())
```

### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Join Plot

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```
